### Plotes mltiples.

```python
g1=sns.FacetGrid(data=propinas,col='time',row='sex')
```

```python
g1=sns.FacetGrid(data=propinas,col='time',row='sex')
g1.map(sns.scatterplot,'total_bill','tip')
```

```python
g2=sns.PairGrid(data=propinas)
```

```python
g2=sns.PairGrid(data=propinas)
g2.map(sns.scatterplot)
```

```python
g2=sns.PairGrid(data=propinas)
g2.map_diag(sns.histplot)
g2.map_offdiag(sns.scatterplot)
```

```python
g3=sns.JointGrid(data=propinas, x='total_bill',y='tip')
```

```python
g3=sns.JointGrid(data=propinas, x='total_bill',y='tip')
g3.plot(sns.scatterplot,sns.histplot)
```

```python
g3=sns.JointGrid(data=propinas, x='total_bill',y='tip')
g3.plot_joint(sns.scatterplot)
g3.plot_marginals(sns.histplot)
```
